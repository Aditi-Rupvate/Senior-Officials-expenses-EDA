# Challenges with the dataset

It has been found that there are several issues with the dataset following an in-depth diagnosis that need to be corrected before any analysis can be conducted. These vary from individual to combined problems across different columns.

\

1. Non-atomic Values :

 Many columns contain non-atomic values, which are entries consisting of more than one piece of information. For example, the "official name" column contains first and last names in one place so that we can replace it with two columns for clearer analysis. The same applies to the "destination" column whereby both city and country reside in one entry, and must be separated for detailed analysis.

 \



2. Missing Values :

 The incomplete entries and null values in specific columns hinder the precision and thoroughness of an analysis. In that case, missing data should be handled in the right way according to the typology and amount of missing data, for example, nullification of cells or imputation techniques replacement or possible removal of partial rows or entire columns from investigation or any other accepted ways.



3. Combination of Both Issues :

 Non-atomic values and missing data appear in some columns at the same time: for example, a city-country combination field may have some missing values for some of its entries. Therefore, those columns will need further preprocessing steps while solving both problems simultaneously since the dataset must be clean for true analyses.

 \

 These individually or collectively prepare the dataset. It will therefore be easier to clean it by separating combined fields, dealing with missing values, and ensuring data integrity. Subsequently, quality will be improved in the dataset, which can make it even more accurate based on further analyses.

1. Non-Atomic Values:

 A column will be broken down into multiple atomic values if there are non-pure values in it - for instance lists - or by combining multiple entries or categories. Thus, each meaningful element is represented in a single column.

 \

2. Missing Values:

Handling nan values : For column modes_of_transport

Find the corresponding destination where modes_of_transport is null, and then for that destination find the highest frequency of modes_of_transport, and this value replaces the null value in modes_of_transport.

```python
# Find the most frequent mode(s)_of_transport for each destination
destination_modes = data.groupby('destination')['modes_of_transport'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)

# Apply logic to fill NaN in 'modes_of_transport' based on corresponding destination's mode
def fill_mode(row):
    if pd.isna(row['modes_of_transport']) and row['destination'] in destination_modes.index:
        return destination_modes[row['destination']]
    return row['modes_of_transport']

# Apply the fill_mode function to each row
data['modes_of_transport'] = data.apply(fill_mode, axis=1)

# Display the updated dataframe
print(data[['destination', 'modes_of_transport']].head())
```

```python
# Step 1: Define a function to handle outlier replacement
def replace_outliers_with_mean(column):
    while True:
        # Calculate Q1, Q3, and IQR
        Q1 = column.quantile(0.25)
        Q3 = column.quantile(0.75)
        IQR = Q3 - Q1

        # Calculate fixed lower and upper bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify non-outliers for mean calculation
        non_outliers = column[(column >= lower_bound) & (column <= upper_bound)]
        mean_without_outliers = non_outliers.mean()

        # Check for outliers
        outliers = column[(column < lower_bound) | (column > upper_bound)]
        if outliers.empty:
            break  # Exit loop if no more outliers

        # Replace outliers with the mean
        column = column.apply(lambda x: mean_without_outliers if x < lower_bound or x > upper_bound else x)

    return column

# Step 2: Apply the function to the desired column
data['sub_total_of_all_transport_used'] = replace_outliers_with_mean(data['sub_total_of_all_transport_used'])
```
